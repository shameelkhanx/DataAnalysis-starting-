```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(1,5,11)
```

```python
y = x**2

```

```python
x
```

```python
y
```

```python
#FunctionalMethod
```

```python
plt.plot(x,y,'y-')
plt.xlabel('LAMBDA(METER)')
plt.ylabel('TEMPERATURE(KELVIN)')
plt.title('WEINS LAW')
```

```python
plt.show()
```

```python
plt.subplot(1,1,1)
plt.plot(x,y,'b-')
```

```python
#object oriented
```

```python
fig = plt.figure()
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
axis.set_xlabel('LAMBDA(m)')
axis.set_ylabel('Temperature(k)')
axis.set_title('Weins Law');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x,'r')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('BIG')
axes2.plot(y,x,'r')
axes2.set_title('SMALL')
plt.show()
```
